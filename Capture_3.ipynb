# 数据结构分类
- 线性：大多常用的
- 非线性：树、堆、哈希表、元素
    - 树形：树、堆、哈希表
    - 网状：图
 

 **所有的数据结构都是基于链表和数组的组合实现的**

# 基本数据类型
CPU可以直接进行运算的类型
- 整数：byte、short、int、long
- 浮点数：float、double
- 字符：char
- 布尔：bool

基本数据类型以二进制的形式存储在计算机中。一个二进制位即为1比特。1字节由8比特组成

# 数字编码
数字是以“补码”的形式存储在计算机中的。
- 原码：二进制的最高位视为符号位，0表整数，1表负数，其余位表示数字的值。
- 反码：正数的反码等于其原码，负数的反码是对其原码除符号位外所有位取反。
- 补码：正数的补码与其原码相同，负数的补码是其反码的基础上加1.



# 字符编码
与数字的存储类似，一个字符对应一个二进制数
